```python
x=[1,2,3,4,5,6,7,8]
y=len(x)
sum=0
prod=1
z=0
while(z<y):
  sum=sum+x[z]
  prod=prod*x[z]
  z=z+1
print("Sum: ", sum)
print("Product: ", prod)
```
